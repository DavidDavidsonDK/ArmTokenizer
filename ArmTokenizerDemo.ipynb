# Evaluate Tokenizer on UD_Armenian-ArmTDP Data

```python
#Evaluate on Train Data 804 sentenses
with open('./UD_Armenian-ArmTDP/hy_armtdp-ud-train.conllu', 'r') as f:
    data = ''.join(f.readlines())
    f.close()
    
sentences = parse(data)
total,good,bad = 0,0,0
tokenizer = Tokenizer()


for sentense in sentences:
    text = sentense.metadata['text'] #original text
    
    tok_list = [re.sub('-|—|–','—',x['form']) for x in sentense if isinstance(x['id'],int)] #avoiding '-' mistakes 
    tok_pred = tokenizer.tokenize(text).tokens()
    for y_true, y_pred in zip(tok_list,tok_pred):
        if y_true == y_pred:
            good +=1
        else:
            bad +=1
            print(y_true , y_pred)
            print('original : ' ,text   )
            print('y_true   : ',tok_list)
            print('y_pred   : ',tok_pred)
            print('*'*115)
        total +=1
print('accuracy  {}%'.format(good*100//total))
```

```python
#Evaluate on Train Data 11349 sentenses
with open('./UD_Armenian-ArmTDP/hy_armtdp-ud-test.conllu', 'r') as f:
    data = ''.join(f.readlines())
    f.close()
    
sentences = parse(data)
total,good,bad = 0,0,0
tokenizer = Tokenizer()
for sentense in sentences:
    text = sentense.metadata['text'] #original text
    
    tok_list = [re.sub('-|—|–','—',x['form']) for x in sentense if isinstance(x['id'],int)] #avoiding '-' mistakes 
    tok_pred = tokenizer.tokenize(text).tokens()  #tokenize
    for y_true, y_pred in zip(tok_list,tok_pred): #compare
        if y_true == y_pred:
            good +=1
        else:
            bad +=1
            print(y_true , y_pred)
            print('original : ' ,text   )
            print('y_true   : ',tok_list)
            print('y_pred   : ',tok_pred)
            print('*'*115)
        total +=1
print('accuracy  {}%'.format(good*100//total))
```
